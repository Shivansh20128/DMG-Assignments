```python
#A3
correlation = data_x_numeric_dataframe.corr()
descending_correlation_series = correlation.unstack().sort_values(ascending=False)
max_pos_val = 0
max_pos_index = 0
min_pos_val = 1
min_pos_index = 0
min_neg_val = 0
min_neg_index = 0
for i in range(len(descending_correlation_series)):
  if(descending_correlation_series.index[i][0] != descending_correlation_series.index[i][1]):
    if(max_pos_val < descending_correlation_series[i]):
      max_pos_val = descending_correlation_series[i]
      max_pos_index = i
    if(min_pos_val > descending_correlation_series[i] and descending_correlation_series[i] > 0):
      min_pos_val = descending_correlation_series[i]
      min_pos_index = i
    if(min_neg_val > descending_correlation_series[i]):
      min_neg_val = descending_correlation_series[i]
      min_neg_index = i
```

```python
print("Max Correlation is {} and features are {} and {}".format(max_pos_val,descending_correlation_series.index[max_pos_index][0],descending_correlation_series.index[max_pos_index][1]))
print("Least Correlaration is {} and features are {} and {}".format(min_pos_val,descending_correlation_series.index[min_pos_index][0],descending_correlation_series.index[min_pos_index][1]))
print("Most Anti-collerated with Correlation being {} are the features {} and {}".format(min_neg_val,descending_correlation_series.index[min_neg_index][0],descending_correlation_series.index[min_neg_index][1]))
```

```python
#A4
from scipy.spatial.distance import mahalanobis
data_x_centered = data_x_numeric - data_x_numeric.mean()
cov_matrix = data_x_centered.T.dot(data_x_centered)/(data_x_centered.shape[0] - 1)
inverse_cov_matrix = np.linalg.inv(cov_matrix)
mahalanobis_arr = []
for i in range(len(data_x_numeric)):
  mahalanobis_arr.append(mahalanobis(data_x_numeric[i],np.mean(data_x_numeric,axis = 0),inverse_cov_matrix))
plt.scatter([i for i in range(1,len(mahalanobis_arr) + 1)], mahalanobis_arr)
```

```python
#A5
mean_vector = data_x_numeric.mean()
dev = data_x_numeric - data_x_numeric.mean()
squared_dev = dev * dev
variance = np.mean(squared_dev)
print("Variance is {}".format(variance))
```

```python
#A6
data_x_centered = data_x_numeric - data_x_numeric.mean()
sample_cov_matrix_inner = data_x_centered.T.dot(data_x_centered)/(data_x_centered.shape[0] - 1)
sample_cov_matrix_outer_product = []
for i in range(len(data_x_centered)):
  if(len(sample_cov_matrix_outer_product) == 0):
    sample_cov_matrix_outer_product =  np.outer(data_x_centered[i], data_x_centered[i])
  else:
    sample_cov_matrix_outer_product += np.outer(data_x_centered[i], data_x_centered[i])
sample_cov_matrix_outer_product = sample_cov_matrix_outer_product/(data_x_centered.shape[0] - 1)
print("Sample cov matrix as calculated by inner product is {}".format(sample_cov_matrix_inner))
print("Sample cov matrix as calculated by outer product is {}".format(sample_cov_matrix_outer_product))
```
